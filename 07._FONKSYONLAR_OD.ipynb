#Fonksyonlar - Metotlar
Farklı programlama dillerinde metotlar olarak bilinen fonksyonlar , nesne tabanlı programlamanın temelidir. "Nesne Tabanlı Programlama (NTP)" İngilizce "Object Oriented Programming (OOP)" olarak bilinir. Bu programlama (NTP) yazılımında tekrar eden veya sonradan  kullanılabilen işlemleri ve değişkenleri saklamayı amaçlar.<br>

**ÖRNEĞİN**<br>
1. Bir listeyi birden fazla kez kullanmak istiyorsunuz diyelim. NTP bunu bir kez yazıp, aşağıdaki kodlarda tekrar çağırabilmemizi sağlar.
2. Bir matematik işlemi yapıyoruz diyelim. Bu işlemi bir fonksyonun içine saklayıp, o fonksyonu çağırarak satırlarca kod yazmamıza engel olur.
3. Bir programda yazdığımız işlemleri başka bir programa yükleyerek veya dosyalarına ekleyerek onu kullanabiliriz.Bu işlemi import ile yaparız.
Tıpkı kütüphane ekler gibi kendi kütüphanelerimizi oluşturabiliriz.
