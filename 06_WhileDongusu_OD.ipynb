# While Döngüsü
While döngüsü, for'dan farklı olarak bir koşulun oluşması durumunda çalışır.Bu koşul,sayıların dışında mantıksal koşulları da çalıştırır. En temelinde if ve for'un karışımı gibidir. Örneğin:

1. Eğer a'nın değeri 40'tan küçük olduğu sürece çalış,
2. Eğer parola True olduğu sürece çalış,
3. Eğer kullanıcı_adı "Haşmet Dayı" olduğu sürece çalış.

While,her zaman bir koşula ihtiyaç duyar!

```python
while(True):
  print("Muhammed")
```

```python
bilgi = input("Lütfen kullanıcı adınızı giriniz:")
while(bilgi != "Haşmet Dayı"):
  print("Kullanıcı Adı Yanlış!")
  bilgi = input("Lütfen kullanıcı adınızı giriniz:")
print("Hoşgeldiniz!")
```
