#DÖNGÜLER (For ve While Döngüsü)
---
Diyelim ki, birden fazla değişkenimiz veya çok fazla kendini tekrar eden işlemimiz var.İşte o zaman döngüler kullanıyoruz.Örneğin;

1. 1'den fazla 1000'e kadar sayıları yazdırmak istiyorsam,
2. Elimde 1000 değişkenlik bir listem var ve yazdırmak istiyorsam,
3. İstediğim bir koşul sağlandığı müddetçe çalıştıracak işlemlerim varsa,
4. Kendini tekrar eden işlemlerim varsa döngüleri kullanırım

Toplam iki farklı döngü vardır. Biri **For**,diğeri ise **while**
döngüsüdür. şimdi **For Döngüsünü** inceleyelim.

```python
#For Döngüsünü Kurmak için yapılması gerekenler oldukça basittir:
for i in range(50,100):
  print(i)
#Her zaman son satırı almaz.O yüzden bir rakam fazla yazarım.
#100 değil 101 yazarım.
#Yukarıdaki döngüyü inceliyelim:
#for....in.....:bir kalıptır.
#i yazan şey bir değişkendir
#range ise bir aralık anlamına gelmektedir.
#Range yerine başka şeyler yazılabilir.
#Bir başka örnek daha yapalım:
ismim="Göktuğ Arı"
for a in ismim:
  print(a)
#Adımın harflerini for döngüsü ile parçalayabilirim.
#Satır satır yazdırabilirim.
```

```python
for i in range(99,101):
  print(i)
# Her zaman son satır almaz. o yüzden bir rakam fazla yazarsın.
# 100 değil 101 yazarım.
# yukarıdaki döngüyü inceleyelim.
# ya for....... in...... : Üstüste bir kalıptır.
# i yazan şey değişkendir.
# range ise aralık anlamına gelir.
# range yerine başka şeyler yazılabilir.
# bir başka örnek daha yapalım.
isimim = "Göktuğ arı"
for a in isimim:
  print(a)
# adımı for döngüsü ile parçalayabilirim
# satır satır yazabilirim.
```

##For Döngüsünde İşlemler

For döngüsünde sadece sayıları art arda yazmalıyız. Aynı zamanda birden çok farklı işlem yapabiliriz.

```python
isim = "Göktuğ Arı "

for degisken in range (1,len(isim)):
  #print(degisken)
  print(isim)
```

```python
sayı_1 = int(int(input("Lütfen döngünün başlayacağı sayıyı yazınız")))
sayı_2 = int(int(input("Lütfen döngünün biteceği ayıyı yazınız")))
for degisken in range(sayı_1,sayı_2):
  print(degisken)
```

```python
for a in range(0,20,2):
  print(a)

  #başlangıç,bitiş,kaçar kaçar gittiğini gösterir
```

**For döngüsünde range () üç değer alabilir. Birincisi başlangıç ikincisi bitişte yeri üçüncü ise döngünün kaçar kaçar artacağını gösterir.**

```python
for a in range (0,50):
  if (a % 5 == 0):
    print(a)
  else:
    print("Bu sayı o sayı değildir!")
```

**NOT:** **% (yüzde)** işareti sayının moduna alma anlamına gelir.
Örneğin bir sayının bölümünde kalanı verir.Mesela "a" sayısının 6'ya
bölünme kalanına mod denir.

```python
for a in range (0,50):
  if (a % 15 == 0):
    print("15'e tam bölünen sayı",a)
  if (a % 5 == 0 or a % 3 == 0):
    print(a)
```

```python
for a in range (0,5):
  print("a değişkeni şunu yazdı:",a)
  for e in range (0,5):
    print("e değişkeni şunu yazdı:",e)
```
